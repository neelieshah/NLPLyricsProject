```python
import pandas as pd
```

```python
df = pd.read_csv('with_lyrics.csv')
```

```python
top_songs = df[df['Peak Position'] <= 10]
```

```python
df['Peak Position'] = df[df['Peak Position'] > 10]
```

```python
plot = df.plot(x = 'time_signature', y='Peak Position', kind = 'scatter')
```

```python
plot = df.plot(x = 'danceability', y='Peak Position', kind = 'scatter')
```

```python
ax = top_songs.plot(x = 'genre', y='spotify_track_duration_ms', kind = 'bar') #top ten = blue
df.plot(x = 'genre', y='spotify_track_duration_ms', kind = 'bar',ax=ax, color = 'red')#regular = red
```
