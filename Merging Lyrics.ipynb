```python
import pandas as pd
```

```python
lyrics = pd.read_csv('lyrics.csv')
```

```python
songs = pd.read_csv('cleaned_data.csv')
```

```python
lyrics['artist'] = lyrics['artist'].replace(['-'], [' '], regex=True)
lyrics['song'] = lyrics['song'].replace(['-'], [' '], regex=True)
```

```python
lyrics['SongID'] = lyrics['song']+lyrics['artist']
```

```python
songs['SongID'] = songs['SongID'].str.lower()
```

```python
merged = pd.merge(songs,lyrics,left_on='SongID',right_on='SongID', how = 'inner')
```
