```python
import pandas as pd
```

```python
lyrics = pd.read_csv('lyrics.csv')
```

```python
songs = pd.read_csv('cleaned_data.csv')
```

```python
songs
```

```python
lyrics['artist'] = lyrics['artist'].replace(['-'], [' '], regex=True)
lyrics['song'] = lyrics['song'].replace(['-'], [' '], regex=True)
```

```python
lyrics
```

```python
lyrics['songID'] = lyrics['song']+lyrics['artist']
```

```python
songs['SongID_caller'] = songs['SongID_caller'].str.lower()
```

```python
merged = songs.join(lyrics, how = 'inner')
```

```python
merged
```

```python
drop_empty = merged.dropna(axis = 0, how = 'any', thresh = None, subset= None, inplace = False) 
```

```python
drop_empty
```

```python
drop_empty['WeekID'].value_counts()
```

```python
drop_empty.to_csv('with_lyrics.csv', index = False)
```
